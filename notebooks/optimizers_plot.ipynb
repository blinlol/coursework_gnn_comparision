```python
import pandas
import seaborn as sns
import matplotlib.pyplot as plt 

optimizers = [
    "adadelta",
    "adagrad",
    "adam",
    "adamax",
    "adamw",
    "asgd",
    "lbfgs",
    "nadam",
    "radam",
    "rmsprop",
    "rprop",
    "sgd",
    "sparseadam"
]

datasets = [
    "Cora",
    "BZR"
]
```

```python
data = None
for opt in optimizers:
    df = pandas.read_csv(f"results_{opt}.csv", index_col='index')
    df['optimizer'] = opt
    data = pandas.concat([data, df], axis=0)
display(data)
df.index.nunique()
```

```python

x = 'test_accuracy'
# x = 'training_time'
y = 'layer'
hue = 'layers_number'
figsize = (8, 4)
for dataset in datasets:
    for opt in optimizers:
        d = data.loc[(data.dataset == dataset) & (data.optimizer == opt)]
        plt.subplots(figsize=figsize)
        plt.title(dataset + ", " + opt)
        sns.boxplot(
            d,
            x=x,
            y=y,
            hue=hue,
            legend=True,
            gap=.2,
            width=0.5
        )

        sns.stripplot(
            d, 
            x=x, 
            y=y, 
            hue=hue, 
            dodge=True, 
            jitter=True,
            palette="pastel"
        )
```
