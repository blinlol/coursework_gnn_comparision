```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("results_adam.csv", index_col="index")
data_mv = pd.read_csv("mean_var.csv", index_col=0)

planetoid = ['Cora', 'CiteSeer', 'PubMed']
tudataset = ['MUTAG', 'PROTEINS', 'COX2', 'BZR', 'AIDS']
```

```python
#       ,    

# x = 'test_accuracy'
x = 'training_time'
y = 'layer'
hue = 'layers_number'
figsize = (8, 4)
for dataset in planetoid + tudataset:
    d = data.loc[(data.dataset == dataset)]
    plt.subplots(figsize=figsize)
    plt.title(dataset)
    sns.boxplot(
        d,
        x=x,
        y=y,
        hue=hue,
        legend=True,
        gap=.2,
        width=0.5
    )

    # sns.stripplot(
    #     d, 
    #     x=x, 
    #     y=y, 
    #     hue=hue, 
    #     dodge=True, 
    #     jitter=True,
    #     palette="pastel"
    # )
```

```python
#      ,       K

for dataset in planetoid + tudataset:
    for layer in data[data.dataset == dataset].layer.unique():
        d = data.query(f'dataset == "{dataset}" and layer == "{layer}"')
        d.fillna(-1, inplace=True)
        d['layers_number'] = d.layers_number.astype(str)
        
        plt.subplots(figsize=figsize)
        plt.title(f"{dataset}, {layer}")
        sns.boxplot(
            d,
            x='test_accuracy',
            y='layers_number',
            hue='K',
            legend=True,
            gap=.2,
            width=0.5
        )


        sns.stripplot(
            d, 
            x='test_accuracy', 
            y='layers_number', 
            hue='K', 
            dodge=True, 
            jitter=True,
            palette="deep"
        )
```
